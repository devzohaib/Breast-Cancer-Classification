```python
from deeppavlov import build_model, configs

model = build_model(configs.squad.squad_ru, download=True)
```

```python

import json

dev_path = '/content/drive/MyDrive/intern Task/RUBQ/RuBQ_2.0_dev.json'
context_path = '/content/drive/MyDrive/intern Task/RUBQ/RuBQ_2.0_paragraphs.json'
test_path = '/content/drive/MyDrive/intern Task/RUBQ/RuBQ_2.0_test.json'

with open(context_path) as f:
  context = json.load(f)

with open(dev_path) as f:
  dev = json.load(f)
  
with open(test_path) as f:
  test = json.load(f)

# creating a dict that map , uid to text in paragraphs.json file
uid_to_text = {}
for i in range(len(context)):
  uid = context[i]['uid']
  text = context[i]['text']
  
  uid_to_text[uid] = text


def Extract_QA(file):
  contexts = []
  questions = []
  answers = []

  for i in range(len(file)):
    question = file[i]['question_text']
    answer_text = file[i]['answer_text']
    uids = file[i]['paragraphs_uids']['with_answer']

    if len(uids) != 0: 
      for uid in uids:

        context = uid_to_text[uid] # get context from uid
        if context.find(answer_text) != -1 :

          # then we find the index of answer in contest
          answer_start = context.index(answer_text)
          answer_end = answer_start + len(answer_text) - 1

          answer= {
              'text':answer_text,
              'answer_start' : answer_start,
              'answer_end' : answer_end
          }

          contexts.append(context)
          questions.append(question)
          answers.append(answer)

  # return pd.DataFrame(data_row)
  return contexts, questions, answers

train_contexts, train_questions, train_answers = Extract_QA(test)
val_contexts, val_questions, val_answers = Extract_QA(dev)
```

```python
train_answers[0]['text']
```

```python
import editdistance
edit_distance = []

for context,question,answer in zip(val_contexts,val_questions,val_answers):
  contxt = [context]
  ques = [question]
  actual_ans = answer['text']

  pred = model(contxt, ques)

  prediction = pred[0][0]
  
  e_dist = editdistance.eval(prediction,actual_ans)
  edit_distance.append(e_dist)

  print("Actual Ans :",actual_ans)
  print("Pred   Ans :",prediction , " ED: ",e_dist)

  print("----")

```

```python
print("total Q : ",len(edit_distance))
```

```python
dis_0 = 0
dis_1 = 0
dis_2 = 0
dis_3 = 0
dis_4 = 0

for dis in edit_distance:
  if dis == 0:
    dis_0 += 1

  elif dis == 1:
    dis_1 +=1
  
  elif dis == 2:
    dis_2 +=1
  
  elif dis == 3:
    dis_3 += 1
  
  elif dis == 4:
    dis_4 += 1


print(dis_0)
print(dis_1)
print(dis_2)
print(dis_3)
print(dis_4)


 
```
